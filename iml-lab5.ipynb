```python
#216270307016
#Dhruti Bhundiya
# to find the maximum and minimum value of a given flattened array
import numpy as np 


arr = np.array([[18,34,74],[25,56,84]])

print("Original array :- ",end="")
print(arr)

print("Maximum value :- ",end="")
max = np.amax(arr)
print(max)

print("Minimum value :- ",end="")
min = np.amin(arr)
print(min)
```

```python
#216270307016
#Dhruti Bhundiya
#to compute the mean , standard devitation , and variance of a given array along the second axis
import numpy as np

arr = np.array([[10,20,30],[40,50,60]])

print("Original array :- ",end="")
print(arr)

mean = np.mean(arr, axis = 1)
print("Mean value :-",end="")
print(mean)

std = np.std(arr, axis = 1)
print("Standar deviation :- ",end="")
print(std)

var = np.var(arr, axis = 1)
print("Variance :- ",end="")
print(var)
```
