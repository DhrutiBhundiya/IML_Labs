```python
#216270307016
#Dhruti Bhunidya
# to add,subtract,multiply , divide argument  element-wise
import numpy as np
x = np.arange(4)
print("x =", x)
print("x + 5 =", x + 5)      #addition
print("x - 5 =", x - 5)      #substrtion
print("x * 2 =", x * 2)      #Multiplication
print("x / 2 =", x / 2)       #division



```

```python
#216270307016
#Dhruti Bhunidya
# to round elements of the array to the nearest integer
import numpy as np
x=np.round([.5, 1.5, 2.5, 3.5, 4.6, 5.4, 6.5, 7.8])
print("Nearest round integer value",x)

```

```python
#216270307016
#Dhruti Bhunidya
#to calculate mean across dimension , in 2d numpy array
import numpy as np 
x=np.array([[5,5],[3,3]])
print("x : \n",x)
print("Mean of x :",np.mean(x))
print("Mean of a along axis=0 :",np.mean(x,axis=0))
print("Mean of a along axis=1 :",np.mean(x,axis=1))

```

```python
#216270307016
#Dhruti Bhunidya
#to calculate the diffrence between neighboring element , element wise of a given array
import numpy as np

arr = np.array([5,9,25,50,75,95,125,150])
print("Original array :- ",end="")
print(arr)

ans = np.diff(arr)
print("Ans :- ",end="")
print(ans)
```
